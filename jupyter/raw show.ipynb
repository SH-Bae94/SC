```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import io
import scipy
import pylab
from scipy.fftpack import fft
from scipy import pi
import math
from pylab import *
from numpy.fft import fft, fftshift
```

```python
%matplotlib inline
```

```python
data = pd.read_excel('Amino4.xlsx')
data.head()
```

```python
plt.figure(figsize=(8,4))
plt.plot(data.time, data.value, label='value')
plt.plot(data.time, data.group, label='group')
plt.grid()
plt.legend()
plt.show()
plt.savefig("figure1.png")
```

FFT 시도

```python
import math
from pylab import *
from numpy.fft import fft, fftshift
```

```python
Fs = 10500.
Ts = 1./Fs

N = 715
freqstep = Fs/N

f = 10*freqstep
t = data.time
y = data.value
```

```python
Y = fft(y)
Y = fftshift(Y)
```

```python
figure(figsize = (8,4))
subplots_adjust(hspace=.4)
```

```python
plot(t,y, '.-')
grid("on")
xlabel('Time (seconds)')
ylabel('Amplitude')
title('signals')
axis('tight')

freq = freqstep * arange(-N/2, N/2)

plt.show()
plt.savefig("Amplitude1.png")
```

```python
plot(freq,abs(Y), '.-')
grid("on")
xlabel('Frequency')
ylabel('Magnitude (Linear)')

plt.show()
plt.savefig("Freq-Magnitude1.png")
```

```python
plot(t,abs(Y), '.-')
grid("on")
xlabel('Time')
ylabel('Magnitude (Linear)')

plt.show()
plt.savefig("Time-Magnitude1.png")
```

```python
plot(freq, angle(Y), '-')
grid("on")
xlabel('Frequency')
ylabel('Phase (Radian)')

plt.show()
plt.savefig("Freq-Phase1.png")
```

```python
plot(t, angle(Y), '-')
grid("on")
xlabel('Time')
ylabel('Phase (Radian)')

plt.show()
plt.savefig("Time-Phase1")
```
