```python
import math
from pylab import *
from numpy.fft import fft, fftshift
```

```python
Fs = 2000.
Ts = 1./Fs

N = 200
freqstep = Fs/N

f = 10*freqstep
t = data.time
y = data.value
```

```python
Y = fft(y)
Y = fftshift(Y)
```

```python
figure(figsize = (8,4))
subplots_adjust(hspace=.4)
```

```python
plot(t,y, '.-')
grid("on")
xlabel('Time (seconds)')
ylabel('Amplitude')
title('signals')
axis('tight')

freq = freqstep * arange(-N/2, N/2)

plt.show()
plt.savefig("Amplitude1.png")
```

```python
plot(freq,abs(Y), '.-')
grid("on")
xlabel('Frequency')
ylabel('Magnitude (Linear)')

plt.show()
plt.savefig("Freq-Magnitude1.png")
```

```python
plot(t,abs(Y), '.-')
grid("on")
xlabel('Time')
ylabel('Magnitude (Linear)')

plt.show()
plt.savefig("Time-Magnitude1.png")
```

```python
plot(freq, angle(Y), '-')
grid("on")
xlabel('Frequency')
ylabel('Phase (Radian)')

plt.show()
plt.savefig("Freq-Phase1.png")
```

```python
plot(t, angle(Y), '-')
grid("on")
xlabel('Time')
ylabel('Phase (Radian)')

plt.show()
plt.savefig("Time-Phase1")
```

```python
frequencies = np.arange(0.0, 30200, 1)
samplingFrequency = 400
```

```python
s1 = np.empty([0])
s2 = np.empty([0])
```

```python
start = 1
```

```python
stop = samplingFrequency + 1
```
